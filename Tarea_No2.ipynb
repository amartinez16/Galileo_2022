#**GIT**#

Es una herramienta para el control de versionamiento que nos ayuda en el seguimiento de los cambios efectuados en nuestro código fuente, así como en colecciones de archivos y carpetas, Git nos solo es para personas que trabajan individualmente es totalmente para el uso compartido y colaborativo.


Git nos permite ver Snapshots de proyecto que hemos trabajados en el pasado, además nos permite ver lo que otras personas han cambiado y nos ayuda en resolver conflictos en la concurrencia dentro de un equipo de desarrollo.


Git mantiene un directorio de nivel superior en donde se almacena las Snapshots, dentro de git conocemos los archivos como Blob y el directorio como tree que a la vez un tree puede tener en su interior otro “tree” y sus “blob”.


**Punto para resaltar en Git**:
* Tiene comandos para poder ser utilizados.
* Todas sus Snapshots están identificadas con punteros referenciados hacia tablas hashes Sha1.
* Los repositorios contienen objetos y referencias.
* Se pueden crear commit para confirmar cada Snapshots de blob modificados creando una estructura o copia donde se especifica cuando se creó, el autor y el mensaje.
*	Nos permite regresar a versiones anteriores de nuestros objetos.





##Ejemplo crear un seguimiento de versionamiento de un archivo

Abrimos nuestra terminal y ejecutamos GIT Bash
1.PNG

Usaremos como ejemplo una carpeta llamada "Ejemplo" que tenemos en el escritorio y que en su interior tienen un archivo .txt llamado "Información".


Cuando llegamos a la carpeta destino ejecutamos el comando git init para inicializar y crear un seguimiento.

Luego si ejecutamos git status se nos mostrara un mensaje que nos muestra que no hay un commit para los archivos de la carpeta y que el archivo esta untracked.

Si deseamos añadir un archivo que esta dentro de la carpeta para iniciar su seguimiento de versionamiento usamos el comando git add  <nombrearchivo>

Luego si ejecutamos otra vez el comando gis status se nos mostrara que ya hay un archivo pero sin commit

Pare realizar nuestra primara instantánea de la versión inicial del archivo y que se almacene en el historial, y poder regresar a ella ante cualquier modificación mal realizada, usaremos el comando git commit donde se mostrara una pantalla para ingresar el mensaje del cambio a realizar.

Y con estos pasos ya tenemos listo nuestro archivo para que git realice un seguimiento de todos los cambios realizados y poder guardar las instantáneos con el comando git commit y poder regresar en algún momento del tiempo en alguna instantánea guardada.
